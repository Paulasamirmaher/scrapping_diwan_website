```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
url='https://diwanegypt.com/product-category/books/arabic-books/'
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
with open('Arabic_books.csv','w',encoding='UTF-8') as f:
    writer=csv.DictWriter(f, fieldnames=['title','author_name','price'])
    writer.writeheader()
    page=soup.find_all('a',{'class':'page-numbers'})[-2]
    lastpage=int(page.get_text())
    for i in range(1,lastpage):
        newurl=f'https://diwanegypt.com/product-category/books/arabic-books/page/{i}/'
        response2=requests.get(newurl)
        soup2=BeautifulSoup(response2.text,'html.parser')
        books=soup2.find_all('a',{'class':"woocommerce-LoopProduct-link woocommerce-loop-product__link"})
        for i in range(len(books)):
            title=soup2.find_all('h2',{'class':"woocommerce-loop-product__title"})[i].get_text()
            nameofauthor=soup2.find_all('span',{'class':'author'})[i].get_text()
            price=soup2.find_all('span',{'class':'price'})[i].get_text()[3:]
            writer.writerow({'title':title,'author_name':nameofauthor,'price':price})
            
```
